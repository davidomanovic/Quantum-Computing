## Circuit library in QiSkit

Standard gates. First we can create a circuit with a Hadamard gate and a multi-controlled-X gate.

```python
from qiskit import QuantumCircuit
from qiskit.circuit.library import HGate, MCXGate
mcx_gate = MCXGate(3) # Multi-controlled-X gate
hadamard_gate = HGate()

qc = QuantumCircuit(4)
qc.append(hadamard_gate, [0])
qc.append(mcx_gate, [0, 1, 2, 3])
qc.draw('mpl')
```

## N-local circuits

We build circuits that alternate layers of single-qubit rotation gates with layers of multi-qubit entangling gates.

```python
from qiskit.circuit.library import TwoLocal
two_local = TwoLocal(3, 'rx', 'cz')
two_local.decompose().draw('mpl')
```

```python
bound_circuit = two_local.assign_parameters({p: 0 for p in two_local.parameters})
bound_circuit.decompose().draw('mpl')
```
